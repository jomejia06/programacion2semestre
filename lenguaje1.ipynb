```
import matplotlib.pyplot as ptl
import cv2 
x = [1,2,3,4,5,6,7,8,9]
y1 = [1,3,5,3,1,3,5,3,1]
y2= [2,4,6,4,2,4,6,4,2]
ptl.plot(x, y1, label="line 1")
ptl.plot(x, y2, label="line 2")

ptl.xlabel("x axis")
ptl.ylabel("y axis")
ptl.title("ejemplo de una grafica lineal")
ptl.legend()
ptl.show()
```

```
import matplotlib.pyplot as plt

# Look at index 4 and 6, which demonstrate overlapping cases.
x1 = [1, 3, 5, 5.5, 9, 10, 9]
y1 = [8, 7, 6, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

# Colors: https://matplotlib.org/api/colors_api.html

plt.bar(x1, y1, label="red Bar", color='r')
plt.bar(x2, y2, label="black Bar", color='b')
plt.plot()

plt.xlabel("linea en x")
plt.ylabel("linea en y ")
plt.title("grafico de barras")
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Use numpy to generate a bunch of random data in a bell curve around 5.
n = 5 + np.random.randn(1000)

m = [m for m in range(len(n))]
plt.bar(m, n)
plt.title("Raw Data")
plt.show()

plt.hist(n, bins=20)
plt.title("Histogram")
plt.show()

plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```

```
import matplotlib.pyplot as plt

x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 4]
y3 = [6, 8, 7, 8, 7]

# Markers: https://matplotlib.org/api/markers_api.html

plt.scatter(x1, y1, marker='v')
plt.scatter(x2, y2, marker='+', color='red')
plt.scatter(x2, y3, marker='^', color='m')
plt.title('Scatter Plot Example')
plt.show()
```
